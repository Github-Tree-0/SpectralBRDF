```python
# Generate the sphere distribution
import random

def generate():
    x = random.normalvariate(0, 1)
    y = random.normalvariate(0, 1)
    z = abs(random.normalvariate(0, 1))
    vec = np.array([x, y, z])
    vec = vec / np.sum(vec**2)
    return vec
```

```python
wis = []
for i in range(200):
    wis.append(generate())
wis = np.array(wis)
np.save('wis.npy', wis)

wos = []
for i in range(200):
    wos.append(generate())
wos = np.array(wos)
np.save('wos.npy', wos)
```
